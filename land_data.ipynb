```python
import pandas as pd
import matplotlib.pyplot as plt

import plotly as px
```

```python
df_original = pd.read_csv("/content/landportal-DP-MOD.csv",error_bad_lines=False)
```

```python
df_original.head(10)
```

```python
df_original[["indicatorID","indicatorLabel","country_iso3","countryLabel","year","value"]] = df_original['indicatorID'].str.split(';', expand=True)
```

```python
df_original[['project_info','stage']] = df_original['indicatorLabel'].str.split('-', n=1, expand=True)
```

```python
df_original.head()
```

```python
df_original.drop(columns= "Unnamed: 1", inplace=True)
```

```python
df_original.drop(columns= "Unnamed: 2", inplace=True)
```

```python
df_original.drop(columns= "indicatorLabel", inplace=True)
```

```python
df_original.head(10)
```

```python
df_original.info()
```

```python
df= pd.DataFrame(df_original)
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df= df.apply(lambda s:s.str.replace('"', ""))
df.head()
```

```python
df["year"] = pd.to_datetime(df["year"])
```

```python
df['value'] = df['value'].astype(float)
```

```python
df.info()
```

```python
df['year'] = df['year'].dt.year
```
